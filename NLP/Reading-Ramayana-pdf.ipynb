```python
import io
from pdfminer.high_level import extract_text

def extract_pdf_text(file_path):
    """Extracts text from a PDF file, handling potential Devanagari characters."""
    with open(file_path, 'rb') as pdf_file:
        text = extract_text(pdf_file, codec='ISCII-DEV')  # Specify UTF-8, ISO-8859-1, windows-1256, Devanagari-7 encoding
    return text

# Example usage:
pdf_text = extract_pdf_text(hindipdf)
print(pdf_text)
```

```python
import PyPDF2

def extract_text_with_encodings(pdf_file_path, encodings_to_try=("utf-8", "devanagari-7", "iscii-dev", "iso-8859-1")):
    """Extracts text from a PDF, attempting different encodings."""

    with open(pdf_file_path, "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(pdf_reader.pages)

        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]

            for encoding in encodings_to_try:
                try:
                    content = page.extract_text(codec=encoding)
                    print(f"Encoding: {encoding} - Content:\n{content}")
                    return content  # Return the successfully extracted text
                except Exception as e:
                    print(f"Encoding {encoding} failed: {e}")

            # If all encodings fail, raise an exception
            raise ValueError("Unable to extract text with any of the provided encodings.")

# Example usage:
extracted_text = extract_text_with_encodings(hindipdf)
print("Final extracted text:\n", extracted_text)
```
