# Roman to Integer


```python
# This function returns value of
# each Roman symbol
def value(r):
    if (r == 'I'):
        return 1
    if (r == 'V'):
        return 5
    if (r == 'X'):
        return 10
    if (r == 'L'):
        return 50
    if (r == 'C'):
        return 100
    if (r == 'D'):
        return 500
    if (r == 'M'):
        return 1000
    return -1

def romanToDecimal(txt):
    txt = txt.replace('.','')
    res = 0
    i = 0

    while (i < len(txt)):

        # Getting value of symbol s[i]
        s1 = value(txt[i])

        if (i + 1 < len(txt)):

            # Getting value of symbol s[i + 1]
            s2 = value(txt[i + 1])

            # Comparing both values
            if (s1 >= s2):

                # Value of current symbol is greater
                # or equal to the next symbol
                res = res + s1
                i = i + 1
            else:

                # Value of current symbol is greater
                # or equal to the next symbol
                res = res + s2 - s1
                i = i + 2
        else:
            res = res + s1
            i = i + 1

    return f"{res:03d}"

# Driver code
print("Integer form of Roman Numeral is"),
print(romanToDecimal("MCMIV"))
```

```python
import fitz

pdf_document = fitz.open(pdffile_name)
num_pages = pdf_document.page_count


with open(textfile_name, "w") as file1:
  for page_num in range(num_pages):
    page = pdf_document.load_page(page_num)
    content = page.get_text()

    pagelines = []
    for line in content.split('\n'):
      #line = line.strip()
      if " continued)" not in line:
        pagelines.append(line)


    content = "\n".join(pagelines)

    # Remove extra linebreaks. and extra space

    content = content.replace('\n \n','~').replace('\n',' ').replace('~','\r\n\r\n').replace("  "," ")
    #content = content + f"\n|Page {page_num+1}|\n"

    file1.write(content)

    print(f"Page {page_num + 1} Saved")
```
