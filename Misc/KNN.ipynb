```python
from  sklearn.cluster import KMeans
import pandas as pd
from sklearn import metrics 
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("clean_train_data.csv")
X_train = df.drop(columns=["Response"])
y_train = df.Response

sc = StandardScaler()
X_train = pd.DataFrame(sc.fit_transform(X_train), columns=X_train.columns, index=X_train.index)
```

```python
knn =KMeans(n_clusters=8, max_iter=50, random_state=100)
knn.fit(X_train,y_train)
y_train_pred = knn.labels_
##or
#y_train_pred= knn.predict(X_train1)
```

```python
metrics.accuracy_score(y_train,y_train_pred)
```

```python
pd.DataFrame({"Id":X_train.Id,"Act Response":y_train,"Pred Response":y_train_pred}, index=X_train.index)
```
