```python
def get_href_of_website(url):
  from bs4 import BeautifulSoup
  import urllib.request
  import re

  html_page = urllib.request.urlopen(url)
  soup = BeautifulSoup(html_page)
  all_hrefs = []
  for link in soup.findAll('a'):
    href = link.get('href')
    try:
      if "http" in href:
        #print (href )
        if "?share" not in href and "mailto:" not in href:
          all_hrefs.append(href)
        else:
          pass
      else:
        pass
        #print ("Nothing")
    except:
      pass
  #print (all_hrefs)
  all_hrefs = list(set(all_hrefs))
  all_hrefs.sort()
  return all_hrefs
```
