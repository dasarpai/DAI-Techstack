```python
import numpy as np

# Create a word-word co-occurrence matrix
word_co_occurrences = np.array([
    [1, 2, 3],
    [2, 1, 4],
    [3, 4, 1]
])

# Perform SVD
U, S, Vh = np.linalg.svd(word_co_occurrences)

# Truncate the singular values
S_truncated = S[:2]

# Extract the word embeddings and context embeddings
word_embeddings = U[:, :2]
context_embeddings = Vh[:, :2]

# Print the word embeddings and context embeddings
print(word_embeddings)
print(context_embeddings)
```
