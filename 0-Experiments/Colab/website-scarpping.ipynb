```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://indiankanoon.org/doc/'
num_pages = 2

for i in range(1, num_pages+1):
    url = base_url + str(i) + '/'
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract the content
        content = soup.find('div', class_='content')
        print (i, "::")
        print(content.get_text())

        # Extract href, if any
        href = soup.find('a', href=True)
        if href:
            print('href:', href['href'])

        print('------------------')
    else:
        print('Failed to retrieve page:', url)
```

# request

```python
import requests
from lxml import html

base_url = 'https://indiankanoon.org/doc/'
num_pages = 3

for i in range(2, num_pages+1):
    url = base_url + str(i) + '/'
    response = requests.get(url)

    if response.status_code == 200:
        tree = html.fromstring(response.content)

        # Extract the content
        content = tree.xpath('//div[@class="content"]/text()')
        print('Content:', ''.join(content))

        # Extract href, if any
        href = tree.xpath('//a[@href]/@href')
        if href:
            print('href:', href[0])

        print('------------------')
    else:
        print('Failed to retrieve page:', url)
```
