```
import requests  
file_url = "https://s3-us-west-2.amazonaws.com/allennlp/models/srl-model-2018.05.25.tar.gz"
    
r = requests.get(file_url, stream = True)  
  
with open("/content/gdrive/My Drive/glue/srl-model-2018.05.25.tar.gz", "wb") as file:  
    for block in r.iter_content(chunk_size = 1024): 
         if block:  
             file.write(block)  

```
