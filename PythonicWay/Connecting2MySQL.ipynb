```python
import mysql.connector
import pandas as pd
```

```python
def getMySQLData(db,tab,sys=False):
    try:
        connection = mysql.connector.connect(host='localhost',
                                             database=db,
                                             user='root',
                                             password='Simple@12.')
        if connection.is_connected():
            db_Info = connection.get_server_info()
            print("Connected to MySQL Server version ", db_Info)
            cursor = connection.cursor()
            cursor.execute("select database();")
            record = cursor.fetchone()
            print("Your connected to database: ", record)
            if sys:
                df= pd.read_sql("SELECT table_name FROM information_schema."+ tab+" WHERE table_schema = '"+ db +"';",connection)
                print ("SELECT * FROM information_schema." +tab +" WHERE table_schema = "+ db +";")
            else:
                df= pd.read_sql("SELECT * FROM "+db+"."+tab+";",connection)
            return df


    except Error as e:
        print("Error while connecting to MySQL", e)
    finally:
        if (connection.is_connected()):
            cursor.close()
            connection.close()
            print("MySQL connection is closed")
```
