```python
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds

# Load the 'plant_leaves' dataset
dataset_name = 'bee_dataset' #'plant_leaves'
ds, ds_info = tfds.load(name=dataset_name, split='train', with_info=True)




```

```python
# Get the list of images and their corresponding labels
images = []
labels = []
for data in tfds.as_numpy(ds):
    images.append(data['input'])
    labels.append(data['output'])
```

```python
# Display the images
num_images_to_display = 5
plt.figure(figsize=(12, 6))

for i in range(num_images_to_display):
    plt.subplot(num_images_to_display,1, i + 1)
    plt.imshow(images[i])
    plt.title("Label: " + str(labels[i]))
    plt.axis("off")

plt.show()
```

```python
# List all available datasets in TensorFlow Datasets
all_datasets = tfds.list_builders()

# Check if each dataset contains an 'image' feature
image_datasets = []

for dataset_name in all_datasets:
    # Load dataset information
    try:
      ds_info = tfds.builder(dataset_name).info

      # Check if the dataset contains an 'image' feature
      if 'image' in ds_info.features:
          image_datasets.append(dataset_name)
    except:
      continue

# Print the names of datasets that contain images
print("Datasets with images:")
print(image_datasets)
```
