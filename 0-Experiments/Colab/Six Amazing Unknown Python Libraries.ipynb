# pendulum

```python
!pip install pendulum
```

```python
# import library

import pendulum
dt = pendulum.datetime(2023, 1, 31)
print(dt)

#local() creates datetime instance with local timezone

local = pendulum.local(2023, 1, 31)
print("Local Time:", local)
print("Local Time Zone:", local.timezone.name)

# Printing UTC time

utc = pendulum.now('UTC')
print("Current UTC time:", utc)

# Converting UTC timezone into Europe/Paris time

europe = utc.in_timezone('Europe/Paris')
print("Current time in Paris:", europe)

india = utc.in_timezone('Asia/Calcutta')
print("Current time in India:", india)
```

```python
# Importing libraries
import sketch
import pandas as pd
# Reading the data (using twitter data as an example)
df = pd.read_csv("/content/sample_data/california_housing_test.csv")
print(df)
```

```python
# Asking to provide code snipped for visualising the emotions
df.sketch.howto("Visualize the emotions")
```

```python
import matplotlib.pyplot as plt

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the data
ax.scatter(df['longitude'], df['latitude'], c=df['median_house_value'], cmap='coolwarm', s=50)

# Set the x and y labels
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

# Set the title of the plot
ax.set_title('Median House Value by Location')

# Show the plot
plt.show()
```

# pgeocode

```python
!pip install pgeocode
```

```python
# Checking for country "India"
import pgeocode

nomi = pgeocode.Nominatim('In')

# Getting geo information by passing the postcodes

nomi.query_postal_code(["620018", "620017", "620012", "110001","248001"])
```

```python
# Finding a distance between two postcodes
distance = pgeocode.GeoDistance('In')
distance.query_postal_code("620018", "620012")
```

```python
# Finding a distance between two postcodes
# “pgeocode” calculates the distance between two postcodes by taking the country and postcodes as inputs. The result is given in kilometers.

distance = pgeocode.GeoDistance('In')
distance.query_postal_code("110001", "248001")
```

```python
from rembg import remove
import cv2
# path of input image (my file: image.jpeg)
input_path = '/content/drive/MyDrive/DS-Data/images/Cat03.jpg'
# path for saving output image and saving as a output.jpeg
output_path = 'output.jpeg'
# Reading the input image
input = cv2.imread(input_path)
# Removing background
output = remove(input)
# Saving file
cv2.imwrite(output_path, output)
```
